```python
import pandas as pd
import numpy as np

import seaborn as sns
from matplotlib import pyplot as plt

from IPython.display import display
from sklearn.model_selection import train_test_split
from sklearn.metrics import mutual_info_score
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('./data.csv.1')
df
```

## Data prep

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')

string_columns = list(df.dtypes[df.dtypes == 'object'].index)

# for col in string_columns:
#     df[col] = df[col].str.lower().str.replace(' ', '_')

df = df[[
    'make',
    'model',
    'year',
    'engine_hp',
    'engine_cylinders',
    'transmission_type',
    'vehicle_style',
    'highway_mpg',
    'city_mpg',
    'msrp'
]]

df = df.fillna(0)
df.rename(columns={'msrp':'price'}, inplace=True)

m = df.price.mean()
df['above_average']  = df.price.apply(lambda x: 1 if x > m else 0)

categorical = ['make', 'model', 'transmission_type', 'vehicle_style']
numerical = ['year','engine_hp', 'engine_cylinders', 'highway_mpg', 'city_mpg']

df_train_full, df_test = train_test_split(df, test_size=0.2, random_state=1)
df_train, df_val = train_test_split(df_train_full, test_size=0.25, random_state=1)

price_train = df_train.price.values
price_val = df_val.price.values
price_test = df_test.price.values

y_train = df_train.above_average.values
y_val = df_val.above_average.values
y_test = df_test.above_average.values

del df_train['price']
del df_val['price']
del df_test['price']

len(df_train), len(df_val), len(df_test)
```

# Q1

```python
from sklearn.metrics import roc_auc_score

for n in numerical:
    res = roc_auc_score(y_train, df_train[n])
    if res < 0.5:
        res = roc_auc_score(y_train, -df_train[n])
    print(n, res)
```

# Q2

```python
dv = DictVectorizer(sparse=False)

train_dict = df_train[categorical + numerical].to_dict(orient='records')
val_dict = df_val[categorical + numerical].to_dict(orient='records')

X_train = dv.fit_transform(train_dict)
X_val = dv.transform(val_dict)

y_train = df_train.above_average.values
y_val = df_val.above_average.values
y_test = df_test.above_average.values

model = LogisticRegression(solver='liblinear', C=1.0, max_iter=1000)
model.fit(X_train, y_train)

y_pred = model.predict_proba(X_val)[:, 1]

roc_auc_score(y_val, y_pred)
```

# Q3

```python
def yolo(y_val, y_pred):
    scores = []

#     thresholds = np.linspace(0, 1, 101)
    thresholds = np.arange(0, 1.01, 0.01)


    for t in thresholds:
        actual_positive = (y_val == 1)
        actual_negative = (y_val == 0)

        predict_positive = (y_pred >= t)
        predict_negative = (y_pred < t)

        tp = (predict_positive & actual_positive).sum()
        tn = (predict_negative & actual_negative).sum()

        fp = (predict_positive & actual_negative).sum()
        fn = (predict_negative & actual_positive).sum()

        scores.append((t, tp, fp, fn, tn))

    columns = ['threshold', 'tp', 'fp', 'fn', 'tn']
    df_scores = pd.DataFrame(scores, columns=columns)    
    return df_scores

df_scores = yolo(y_val, y_pred)

df_scores.head()
```

```python
df_scores['tpr'] = df_scores.tp / (df_scores.tp + df_scores.fn)
df_scores['fpr'] = df_scores.fp / (df_scores.fp + df_scores.tn)

df_scores['p'] = df_scores.tp / (df_scores.tp + df_scores.fp)
df_scores['r'] = df_scores.tp / (df_scores.tp + df_scores.fn)

plt.plot(df_scores.threshold, df_scores.p, label='')
plt.plot(df_scores.threshold, df_scores.r, label='FPR')
plt.legend()


```
